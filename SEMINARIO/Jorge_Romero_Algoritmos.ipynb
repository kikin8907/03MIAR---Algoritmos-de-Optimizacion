### (*)¿Cuantas posibilidades hay sin tener en cuenta las restricciones?

Sin tener en cuenta las restricciones obtenemos que esto es una Variacion con Repeticion, puestro podriamos utilizar nuevamente los signos y los numeros y dado que para no todos los signos basicos cumplen con la propiedad conmutativa, entonces cumple la otra caracteristica de la variacion con repeticion que es la *importancia del orden* en como usamos tanto numeros como signos, esto quiere decir.

Para el caso de los signos

$Vr = 4^4 = 256$

Para el caso de los números
$Vr = 9^5 = 59049$

Multiplicando ambas variaciones obtenemos un total de $59049* 256 = 15116544$ posibles variaciones 

###¿Cuantas posibilidades hay teniendo en cuenta todas las restricciones.

Para este caso se observan combinaciones y permutaciones tal como se explica a continuacion.

En el caso de los signos se observaría una permutación.
$P = 4! = 24%$

En el caso de los numeros al tener un grupo de 9 digitos y se podran usar 5 sin repetirse, se observa una combinacion sin repeticion, por lo tanto se observaria: $9!/(5!(9-5)! = 126$

Dentro de la operacion solo podemos usar 5 numeros sin repeticion, por lo tanto obtendriamos una permutacion $5! = 120$

Y finalmente combinando todas las posibles combinaciones obtenemos que
$(126 * 120) *24 = 362880$

###(*)Diseña un algoritmo que mejore la complejidad del algortimo por fuerza bruta. Argumenta porque crees que mejora el algoritmo por fuerza bruta

Respuesta

```
from itertools import permutations, combinations

digitos = (1,2,3,4,5,6,7,8,9)
permu_digitos = [x for x in permutations(digitos, 5)] #### Complejidad computacional de O(n!)

simbolos = ('-','+','/','*')
operaciones = [x for x in permutations(simbolos, 4)] #### Complejidad computacional de O(n!)

def encontrar_operaciones(numero_encontrar):
  soluciones = []

  for digito in permu_digitos:
    for operador in operaciones:
      ### Ensamblaje de operaciones
      combinacion = f'{digito[0]}{operador[0]}{digito[1]}{operador[1]}{digito[2]}{operador[2]}{digito[3]}{operador[3]}{digito[4]}'
      ### Verificaciones sobre la operacion ensamblada
      evaluar_resultado = eval(combinacion)
      if numero_encontrar == evaluar_resultado:
        soluciones.append(combinacion)

  return soluciones

print(f'operaciones que encuentran el digito 8 {encontrar_operaciones(8)}')
```

###(*)Calcula la complejidad del algoritmo 

De acuerdo a una busqueda por internet, las permutaciones usando la libreria itertools, la complejidad computacional es de $O(n!)$ tomando en cuenta los 2 ciclos anidados podriamos decir que nuestra complejidad calculada es de $O(2n!)+ O(n^2)$ que se representaria como $O(n!)$

###Según el problema (y tenga sentido), diseña un juego de datos de entrada aleatorios

```
import random

def datos_aleatorios(x,y):
  random_operaciones = [random.sample(random.choices(simbolos,k=4),4) for a in range(0,x)]
  random_digitos = [random.sample(random.choices(digitos,k=5),5) for b in range(0,y)]
  return random_digitos,random_operaciones

datos_aleatorios(2,2)
```
